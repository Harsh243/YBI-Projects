```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
model=LogisticRegression(max_iter=3000)
from sklearn.metrics import accuracy_score
```

```python
cd = pd.read_csv('/DataSets/creditcard.csv')
```

```python
cd.head()
```

```python
cd.tail()
```

```python
cd.info()
```

```python
cd['Class']=cd['Class'].astype('int')
```

```python
cd.isnull().sum()
```

```python
cd=cd.dropna()
cd.isnull().sum()
```

```python
cd["Class"].value_counts()
```

```python
legit = cd[cd.Class == 0]
fraud = cd[cd.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
cd.groupby("Class").mean()
```

```python
l_s = legit.sample(n=94)
```

```python
nd = pd.concat([l_s, fraud], axis=0)
```

```python
nd.head()
```

```python
nd.tail()
```

```python
nd["Class"].value_counts()
```

```python
nd.groupby('Class').mean()
```

```python
X = nd.drop(columns="Class", axis=1)
Y = nd['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.4, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
X_train_pred = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_pred, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
X_test_pred = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_pred, Y_test)
```

```python
print('Accuracy score on Test Data : ', test_data_accuracy)
```
