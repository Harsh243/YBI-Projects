```python
df['Date'] = df['date'].dt.date
df['Year'] = df['date'].dt.year
df['Month'] = df['date'].dt.month
df['Day'] = df['date'].dt.day
df['Weekday'] = df['date'].dt.day_name()
df['hour'] = df['date'].dt.hour
df.head()
```

```python
def season_creation(x):
    if x in [2,3,4]:
        season = "Spring"
    elif x in [5,6]:
        season = "Summer"
    elif x in [7,8]:
        season = "Monsoon"
    elif x in [9,10]:
        season = "Autumn"
    elif x in[11,12,1]:
        season = "Winter"
    return season
```

```python
df["Season"] = df["Month"].apply(season_creation)
df.head()
```

```python
plt.figure(figsize = (12,7))
sns.boxplot(x = df['Weekday'], y = df['pm2_5'], hue = df['Season'])
plt.show()
```

```python
df_pred = delhi[['Date','pm2_5']]
df_pred.head()
```

```python
df_pred['Date'] = pd.to_datetime(df_pred['Date'])
df_pred.info()
```

```python
df_pred = df_pred.groupby('Date')['pm2_5'].sum().reset_index()
df_pred.head()
```

```python
df_pred1 = df_pred.copy()
df_pred1.set_index('Date', inplace=True)
df_pred1.head()
```

```python
df_pred1.plot()
plt.show()
```

```python
import statsmodels.api as sm
sm.stats.durbin_watson(df_pred1)
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
from pylab import rcParams
rcParams['figure.figsize'] = 12,7
decomposition_additive = seasonal_decompose(df_pred1, model = 'additive', period=12)
decomposition_additive.plot()
plt.show()
```

```python
df_prophet = df_pred.rename(columns={'Date':'ds', 'pm2_5':'y'})
df_prophet
```

```python
from prophet import Prophet
```

```python
m = Prophet()
model = m.fit(df_prophet)
```

```python
model.seasonalities
```

```python
future_data = model.make_future_dataframe(periods = 180, freq='D')
```

```python
future_data
```

```python
df_prophet.shape
```

```python
future_data.shape
```

```python
prediction = model.predict(future_data)
prediction
```

```python
prediction.columns
```

```python
prediction[['ds','yhat_lower', 'yhat_upper', 'yhat']].tail(180)
```

```python
model.plot(prediction)
plt.show()
```

```python
model.plot_components(prediction)
plt.show()
```

```python
from prophet.diagnostics import cross_validation
from prophet.plot import plot_cross_validation_metric
from prophet.diagnostics import performance_metrics
df_cv = cross_validation(model, horizon='30 days', period = '15 days', initial='90 days')
```

```python
df_performance = performance_metrics(df_cv)
df_performance
```

```python
plot_cross_validation_metric(df_cv, metric='rmse')
plt.show()
```
